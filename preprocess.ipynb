```python
import pandas as pd
import numpy as np
import sys
```

```python
crx_train = pd.read_csv('crx.data.training', header=None)
crx_test = pd.read_csv('crx.data.testing', header=None)
```

```python
def process_nu(df):
    nu_columns = [1,2,7,10,13,14]
    df[nu_columns] = df[nu_columns].replace('?', np.NaN)
    df[nu_columns] = df[nu_columns].apply(pd.to_numeric)

    means = df[nu_columns].mean()
    for col in means.index:
        df[col] = df[col].replace(np.NaN, means[col])
    return df
```

```python
def process_bi(df):
    bi_columns = [0,3,4,5,6,8,9,11,12]
    
    for i in bi_columns:
        freqs = df[i].value_counts()  
        df[i] = df[i].replace('?', freqs.index[0])
        
        c = list(df[i].value_counts().index)
        n = list(range(df[i].value_counts().shape[0]))
        dict_ = dict(zip(c,n))
        
        for k,v in df[i].iteritems():
            df.loc[k,i] = dict_[v]
    return df
```

```python
def Z_norm(df):
    length = df.shape[1]
    for i in range(length-1):
        df[i] = (df[i]-df[i].mean())/df[i].std()
    
    return df
```

```python
# df = process_nu(df)
# df = process_bi(df)
# df = Z_norm(df)
def preprocess(df):
    df = process_nu(df)
    df = process_bi(df)
    df = Z_norm(df)
    return df


```

```python
crx_train = preprocess(crx_train)
crx_test = preprocess(crx_test)
```

```python
import pandas as pd
import numpy as np
import sys
import math
```

```python
def distance(a,b):
    d = 0
    for i in range(len(a)):
        d += math.pow((a[i]-b[i]), 2)
    distance = math.sqrt(d)
    return distance
        
```

```python
def knn(testset, trainingset, k):
    testset[testset.shape[1]-1] = np.NaN
    for i, row in testset.iterrows():
        neighbors_ = neighbors(row, trainingset, k)
        label = majority(neighbors_)
        testset.loc[i,testset.shape[1]-1] = label
    return testset
    
```

```python
def neighbors(instance, trainingset, k):
    label_index = trainingset.shape[1]-1
    distances = []
    for i, r in trainingset.iterrows():
        d = distance(list(instance)[0:label_index-1], list(r)[0:label_index-1])
        distances.append((list(r)[label_index], d))
    distances.sort(key=lambda x: x[1])
    k_neighbors = []
    for j in range(k):
        k_neighbors.append(distances[j][0])
    return k_neighbors

def majority(neighbors_):
    return max(neighbors_, key=neighbors_.count)


    
        
```
